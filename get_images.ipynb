```python
from process_data import usar_reportes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mes_a√±o, revenues_per_week,revenues,pizzas_vendidas, revenues_por_pedido,comida_cena, dia_semana, cantidad_pizzas_pedido, pizzas_pedidas= usar_reportes() 
```

```python
fig = plt.figure(figsize=(8,4))
plt.plot(revenues_per_week, color = 'g')
plt.title(f'Revenues per week. Total Revenues: {int(revenues)}')
plt.show()
```

```python

fig, ax1 = plt.subplots(figsize=(8,4))
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
print(comida_cena)
ax1.pie(comida_cena, explode=explode, labels=['Lunch','Diner'], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Probability of ordering by time of meal')
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
sns.barplot(x=mes_a√±o.index, y=mes_a√±o,orient='v',palette='Paired')
plt.title('Probality of ordering by month.')
plt.xticks(rotation=70)
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
sns.barplot(x=dia_semana.index, y=dia_semana,orient='v',palette='viridis')
plt.title('Probality of ordering by day')
plt.xticks(rotation=70)
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
sns.barplot(x=pizzas_pedidas.index, y=pizzas_pedidas,orient='v',palette='rocket')
plt.title('Number of pizzas per type')
plt.xticks(rotation=70)
plt.show()
```

```python
ingredientes = pd.read_csv('compra_semanal_2017.csv')
a = ingredientes.sum().sort_values(ascending = False)[:20]/ingredientes.sum().sum()
fig = plt.figure(figsize=(8,4))
sns.barplot(x=a.index, y=a,orient='v',palette='Paired')
plt.title(f'Most used ingredients. They represent: {int(a.sum()*100)}%')
plt.xticks(rotation=70)
plt.show()
```
